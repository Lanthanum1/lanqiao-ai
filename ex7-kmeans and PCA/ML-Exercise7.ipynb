## K-means 聚类

我们将实施和应用K-means到一个简单的二维数据集，以获得一些直观的工作原理。 K-means是一个迭代的，无监督的聚类算法，将类似的实例组合成簇。 该算法通过猜测每个簇的初始聚类中心开始，然后重复将实例分配给最近的簇，并重新计算该簇的聚类中心。 我们要实现的第一部分是找到数据中每个实例最接近的聚类中心的函数。

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
from scipy.io import loadmat
```

```python
# X: 是一个形状为(m, n)的二维数组，表示有m个样本，每个样本有n个特征。
# centroids: 是一个形状为(k, n)的二维数组，表示有k个聚类中心，每个中心同样具有n个特征。
def find_closest_centroids(X, centroids):
    m = X.shape[0]
    k = centroids.shape[0]
    idx = np.zeros(m)
    #   m: 被赋值为X的行数，即样本数量。
    #   k: 被赋值为centroids的行数，即聚类中心的数量。
    #   idx: 初始化为一个长度为m的全零数组，用于存储每个样本点最近的聚类中心的索引。

    for i in range(m):
        min_dist = 1000000
        for j in range(k):
            dist = np.sum((X[i, :] - centroids[j, :]) ** 2)
            if dist < min_dist:
                min_dist = dist
                idx[i] = j

    return idx
```

让我们来测试这个函数，以确保它的工作正常。 我们将使用练习中提供的测试用例。

```python
data = loadmat("data/ex7data2.mat")
X = data["X"]
initial_centroids = np.array([[3, 3], [6, 2], [8, 5]])

idx = find_closest_centroids(X, initial_centroids)
idx[0:3]
```

```python
data2 = pd.DataFrame(data.get("X"), columns=["X1", "X2"])
data2.head()
```

```python
# k: 表示聚类的数量。
def compute_centroids(X, idx, k):
    m, n = X.shape
    centroids = np.zeros((k, n))

    for i in range(k):
        indices = np.where(idx == i)
        # 对于每个聚类i，使用np.where(idx == i)找到所有分配给该聚类的样本的索引，存储在indices中。

        centroids[i, :] = (np.sum(X[indices, :], axis=1) / len(indices[0])).ravel()
        # 利用得到的索引indices，从X中提取属于该聚类的所有样本点，然后沿轴1（即特征维度）求和，得到一个形状为(1, n)的数组，表示该聚类所有样本的特征总和。
        # 将上述求得的特征总和除以该聚类中样本的数量（len(indices[0])），从而得到该聚类中心的平均值。这里使用.ravel()将可能的多维数组展平为一维，确保赋值给centroids[i, :]时形状匹配。

    return centroids
```

此输出也符合练习中的预期值。 
下一部分涉及实际运行该算法的一些迭代次数和可视化结果。 
这个步骤是由于并不复杂，我将从头开始构建它。 为了运行算法，我们只需要在将样本分配给最近的簇并重新计算簇的聚类中心。

```python
def run_k_means(X, initial_centroids, max_iters):
    m, n = X.shape
    k = initial_centroids.shape[0]
    idx = np.zeros(m)
    centroids = initial_centroids

    for i in range(max_iters):
        idx = find_closest_centroids(X, centroids)
        centroids = compute_centroids(X, idx, k)

    return idx, centroids
```

```python
idx, centroids = run_k_means(X, initial_centroids, 10)
```

我们跳过的一个步骤是初始化聚类中心的过程。 这可以影响算法的收敛。 我们的任务是创建一个选择随机样本并将其用作初始聚类中心的函数。

```python
def init_centroids(X, k):
    m, n = X.shape
    centroids = np.zeros((k, n))
    idx = np.random.randint(0, m, k)
    # 生成一个长度为k的随机整数数组，范围在0到m-1之间。这个数组将用于随机选择X中的k个样本点作为初始聚类中心。

    for i in range(k):
        centroids[i, :] = X[idx[i], :]

    return centroids
```

现在我们需要对数据应用一些预处理，并将其提供给K-means算法。

```python
# normalize value ranges
A = A / 255.0
# 将数组A中的所有元素值进行归一化处理。这里的归一化是将数组中的每个数值除以255，其目的主要是将像素值从0-255的范围转换到0-1的范围。这种操作在图像处理中非常常见，因为许多机器学习算法和深度学习模型期望输入数据在较小的范围内，这样可以提高训练效率，减少梯度消失或爆炸的风险，并使得模型对光照变化等具有更好的鲁棒性。

# reshape the array
X = np.reshape(A, (A.shape[0] * A.shape[1], A.shape[2]))
# 原本的数组A可能是一个三维数组，通常在图像处理中表示为(高度, 宽度, 通道数)。重塑操作将其转换为二维数组，具体形状为(高度 * 宽度, 通道数)。
# A.shape[0]和A.shape[1]分别代表原数组的行数（高度）和列数（宽度）。
# A.shape[2]代表原数组的通道数，对于RGB图像，通道数通常是3。

X.shape
```

```python
# randomly initialize the centroids
initial_centroids = init_centroids(X, 16)
# 随机从数据集X中选择16个样本点作为初始聚类中心。X通常是一个经过预处理（如归一化）的二维数组，每行代表一个特征向量（例如，像素点的颜色值）。选择16个聚类中心意味着我们希望将数据分为16个类别。

# run the algorithm
idx, centroids = run_k_means(X, initial_centroids, 10)
# 调用run_k_means函数执行K-means算法。输入参数包括原始数据集X、初始聚类中心initial_centroids，以及迭代次数10。该函数会反复更新每个样本点所属的聚类（存储在idx中）和聚类中心（存储在centroids中），直到达到指定的迭代次数或满足收敛条件。最终，idx是一个数组，表示每个样本点分配到的聚类索引，而centroids是经过算法优化后的聚类中心。

# get the closest centroids one last time
idx = find_closest_centroids(X, centroids)
# 尽管在run_k_means中已经重复进行了样本到聚类中心的分配，这里再次调用find_closest_centroids是为了确保每个样本点都被分配到了最终确定的聚类中心上。这一步骤在某些情况下可能是为了获得最终的分配结果，特别是在run_k_means的最后一次迭代后，确保所有点都依据最新的聚类中心进行了重新分配。

# map each pixel to the centroid value
X_recovered = centroids[idx.astype(int), :]
# 这一步是将原始数据X中的每个样本点映射到其最近的聚类中心的值上，实现数据的“重构”。通过idx.astype(int)将索引转换为整数（如果有必要的话），然后利用这些索引从centroids中取出对应的聚类中心值，最终得到重构后的数据集X_recovered。

X_recovered.shape
# X_recovered.shape输出重构数据的形状，以验证数据结构是否符合预期，例如，如果X是一个图像数据集，那么X_recovered也应保持相同的维度，只是每个像素点的颜色值被其最近的聚类中心颜色所替代。
```

## Principal component analysis（主成分分析）

PCA是在数据集中找到“主成分”或最大方差方向的线性变换。 它可以用于降维。 在本练习中，我们首先负责实现PCA并将其应用于一个简单的二维数据集，以了解它是如何工作的。 我们从加载和可视化数据集开始。

```python
data = loadmat("data/ex7data1.mat")
data
```

```python
X = data["X"]

fig, ax = plt.subplots(figsize=(12, 8))
ax.scatter(X[:, 0], X[:, 1])
plt.show()
```

PCA的算法相当简单。 在确保数据被归一化之后，输出仅仅是原始数据的协方差矩阵的奇异值分解。

```python
def pca(X):
    # normalize the features
    X = (X - X.mean()) / X.std()

    # compute the covariance matrix
    X = np.matrix(X)
    cov = (X.T * X) / X.shape[0]
    # X.T * X计算转置后的X与自身相乘，这实际上是计算了样本间的协方差。因为X是样本特征的矩阵，所以这个操作的结果是一个对称矩阵，其对角线上的元素表示特征的方差，非对角线元素表示特征间的协方差。
    # (X.T * X) / X.shape[0]将上面的结果除以样本数（X.shape[0]是样本数），得到的是样本协方差矩阵。

    # perform SVD
    U, S, V = np.linalg.svd(cov)
    # NumPy的np.linalg.svd函数对样本协方差矩阵cov进行奇异值分解。这会返回三个矩阵：U（左奇异向量矩阵）、S（对角奇异值矩阵）和V（右奇异向量矩阵）。

    # U的列是主成分，按照特征值的大小排列，其中列向量是正交的。
    # S是对角矩阵，其对角线上的元素是对应的特征值，表示主成分的方差。
    # V是cov的共轭转置的奇异向量矩阵，其列向量与U的列向量正交。

    return U, S, V
```

```python
U, S, V = pca(X)
U, S, V
```

现在我们有主成分（矩阵U），我们可以用这些来将原始数据投影到一个较低维的空间中。 对于这个任务，我们将实现一个计算投影并且仅选择顶部K个分量的函数，有效地减少了维数。

```python
def project_data(X, U, k):
    U_reduced = U[:, :k]
    return np.dot(X, U_reduced)
    # 使用NumPy的np.dot函数进行矩阵乘法，将原始数据集X与降维后的主成分矩阵U_reduced相乘。这一步骤相当于将每个样本沿着前k个主成分的方向进行投影，得到降维后的数据。投影后的结果仍是一个二维数组，其中每一行代表一个样本，但现在每个样本只有k个特征，而不是原始数据集的特征数。
```

```python
Z = project_data(X, U, 1)
Z
```

我们也可以通过反向转换步骤来恢复原始数据。

```python
def recover_data(Z, U, k):
    U_reduced = U[:, :k]
    return np.dot(Z, U_reduced.T)
```

```python
X_recovered = recover_data(Z, U, 1)
X_recovered
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
ax.scatter(list(X_recovered[:, 0]), list(X_recovered[:, 1]))
plt.show()
```

请注意，第一主成分的投影轴基本上是数据集中的对角线。 当我们将数据减少到一个维度时，我们失去了该对角线周围的变化，所以在我们的再现中，一切都沿着该对角线。
